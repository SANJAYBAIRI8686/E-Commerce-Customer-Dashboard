## Data Cleaning

```python
print(" Starting Data Cleaning...\n")

# Store original size
original_size = len(df)

# 1. Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])
print(f" Removed {original_size - len(df):,} rows with missing CustomerID")

# 2. Remove cancelled orders (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print(f" Removed cancelled orders. Current size: {len(df):,}")

# 3. Remove negative quantities and prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print(f" Removed invalid quantities/prices. Current size: {len(df):,}")

# 4. Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(" Converted InvoiceDate to datetime")

# 5. Create TotalAmount column
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
print(" Created TotalAmount column")

# 6. Extract date components
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
print(" Extracted date components")

# 7. Remove outliers (quantities > 99th percentile)
q99 = df['Quantity'].quantile(0.99)
df = df[df['Quantity'] <= q99]
print(f" Removed outliers. Final size: {len(df):,}")

print("\n" + "="*60)
print("DATA CLEANING COMPLETE!")
print("="*60)
print(f"Final dataset: {len(df):,} rows")
print(f"Date range: {df['InvoiceDate'].min()} to {df['InvoiceDate'].max()}")
print(f"Unique customers: {df['CustomerID'].nunique():,}")
print(f"Unique products: {df['StockCode'].nunique():,}")
print(f"Total revenue: ${df['TotalAmount'].sum():,.2f}")
```

## RFM Analysis (Coustomer Segmentation)

```python
print(" Starting RFM Analysis (Recency, Frequency, Monetary)...\n")

# Set analysis date (1 day after last transaction)
analysis_date = df['InvoiceDate'].max() + timedelta(days=1)

# Calculate RFM metrics
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (analysis_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalAmount': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

print("Sample RFM Data:")
print(rfm.head(10))

# Create RFM scores (1-5, where 5 is best)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# Calculate RFM Score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Total'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)

# Customer Segmentation
def segment_customers(row):
    if row['RFM_Total'] >= 13:
        return 'Champions'
    elif row['RFM_Total'] >= 10:
        return 'Loyal Customers'
    elif row['RFM_Total'] >= 7:
        return 'Potential Loyalists'
    elif row['RFM_Total'] >= 5:
        return 'At Risk'
    else:
        return 'Lost'

rfm['Segment'] = rfm.apply(segment_customers, axis=1)

print("\n" + "="*60)
print("CUSTOMER SEGMENTS")
print("="*60)
segment_summary = rfm.groupby('Segment').agg({
    'CustomerID': 'count',
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
}).round(2)
segment_summary.columns = ['Count', 'Avg_Recency', 'Avg_Frequency', 'Avg_Monetary']
print(segment_summary)

# Visualize segments
plt.figure(figsize=(12, 6))
segment_counts = rfm['Segment'].value_counts()
colors = plt.cm.Set3(np.linspace(0, 1, len(segment_counts)))
bars = plt.bar(segment_counts.index, segment_counts.values, color=colors, edgecolor='black')

plt.title('Customer Segmentation Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Segment', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n RFM Analysis Complete!")
```

## Customer Lifetime Value

```python
print(" Calculating Customer Lifetime Value...\n")

# Calculate CLV metrics
clv = rfm.copy()
clv['AvgPurchaseValue'] = clv['Monetary'] / clv['Frequency']
clv['PurchaseFrequency'] = clv['Frequency'] / 365  # Assuming 1 year
clv['CustomerValue'] = clv['AvgPurchaseValue'] * clv['PurchaseFrequency']
clv['CustomerLifespan'] = 365 / clv['Recency']  # Simplified
clv['CLV'] = clv['CustomerValue'] * clv['CustomerLifespan']

# Handle infinities
clv['CLV'] = clv['CLV'].replace([np.inf, -np.inf], np.nan)
clv['CLV'] = clv['CLV'].fillna(clv['CLV'].median())

print("Top 10 Customers by CLV:")
print(clv.nlargest(10, 'CLV')[['CustomerID', 'Segment', 'Monetary', 'Frequency', 'CLV']])

# CLV by Segment
plt.figure(figsize=(12, 7))
segment_clv = clv.groupby('Segment')['CLV'].mean().sort_values(ascending=False)
colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(segment_clv)))

bars = plt.bar(segment_clv.index, segment_clv.values, color=colors, edgecolor='black')
plt.title('Average Customer Lifetime Value by Segment', fontsize=16, fontweight='bold')
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('Average CLV ($)', fontsize=12)
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n Total CLV: ${clv['CLV'].sum():,.2f}")
print(f" Average CLV: ${clv['CLV'].mean():,.2f}")
print(f" Median CLV: ${clv['CLV'].median():,.2f}")
```

## Churn Prediction Model

```python
print(" Building Customer Churn Prediction Model...\n")

# Define churn (customers who haven't purchased in 90+ days)
rfm['Churned'] = (rfm['Recency'] > 90).astype(int)

print(f"Churned customers: {rfm['Churned'].sum()} ({rfm['Churned'].mean()*100:.1f}%)")
print(f"Active customers: {(1-rfm['Churned']).sum()} ({(1-rfm['Churned']).mean()*100:.1f}%)")

# Prepare features
X = rfm[['Recency', 'Frequency', 'Monetary']]
y = rfm['Churned']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nTraining set: {len(X_train)} customers")
print(f"Testing set: {len(X_test)} customers")

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred = rf_model.predict(X_test)

# Evaluation
print("\n" + "="*60)
print("MODEL PERFORMANCE")
print("="*60)
print(f"\nAccuracy: {accuracy_score(y_test, y_pred):.2%}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Active', 'Churned']))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Churn Prediction Confusion Matrix', fontsize=14, fontweight='bold')
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.xticks([0.5, 1.5], ['Active', 'Churned'])
plt.yticks([0.5, 1.5], ['Active', 'Churned'])
plt.tight_layout()
plt.show()

# Feature Importance
feature_importance = pd.DataFrame({
    'Feature': ['Recency', 'Frequency', 'Monetary'],
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.barh(feature_importance['Feature'], feature_importance['Importance'], 
                color=['#ff6b6b', '#4ecdc4', '#45b7d1'], edgecolor='black')
plt.title('Feature Importance for Churn Prediction', fontsize=14, fontweight='bold')
plt.xlabel('Importance Score', fontsize=12)
plt.tight_layout()
plt.show()
```

## Product Association Analysis ( Market Basket)

```python
print(" Performing Market Basket Analysis...\n")

# Create transaction format
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
basket = (basket > 0).astype(int)

# Calculate product co-occurrence
from itertools import combinations

# Get top 20 products
top_20_products = df['Description'].value_counts().head(20).index

# Filter basket
basket_top = basket[top_20_products]

# Calculate co-occurrence matrix
cooccurrence = basket_top.T.dot(basket_top)

# Remove diagonal
np.fill_diagonal(cooccurrence.values, 0)

# Find top associations
associations = []
for i in range(len(cooccurrence)):
    for j in range(i+1, len(cooccurrence)):
        if cooccurrence.iloc[i, j] > 0:
            associations.append({
                'Product_A': cooccurrence.index[i],
                'Product_B': cooccurrence.columns[j],
                'Count': cooccurrence.iloc[i, j]
            })

assoc_df = pd.DataFrame(associations).sort_values('Count', ascending=False)

print("="*60)
print("TOP 10 PRODUCT ASSOCIATIONS")
print("="*60)
print(assoc_df.head(10))

# Visualize top associations
top_assoc = assoc_df.head(10)

plt.figure(figsize=(14, 8))
y_pos = range(len(top_assoc))
labels = [f"{row['Product_A'][:30]}...\n+ {row['Product_B'][:30]}..." 
          for _, row in top_assoc.iterrows()]

bars = plt.barh(y_pos, top_assoc['Count'], color=plt.cm.Paired(np.linspace(0, 1, len(top_assoc))))
plt.yticks(y_pos, labels, fontsize=9)
plt.xlabel('Number of Co-purchases', fontsize=12)
plt.title('Top 10 Product Associations (Frequently Bought Together)', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()

for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f' {int(width)}', 
             ha='left', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n Market basket analysis complete!")
```
